```python
# --- Import necessary modules ---
from lida import Manager, TextGenerationConfig, llm
import os
from IPython.display import display
import traceback  # To print detailed error information

# --- Initialize the LLM with OpenAI (using environment variable for API key) ---
# Make sure you have set the OPENAI_API_KEY environment variable
# (e.g., in your terminal before launching the notebook: export OPENAI_API_KEY="YOUR_KEY")
lida = Manager(text_gen = llm("openai"))

# --- Configure Text Generation ---
textgen_config = TextGenerationConfig(n=1, temperature=0.5, model="gpt-3.5-turbo", use_cache=True)

# --- Load Data and Generate Summary and Goals ---
local_file_path = "cars.csv"  # If you saved it in the same directory as your notebook
summary = lida.summarize(local_file_path, summary_method="default", textgen_config=textgen_config)
print("Data Summary:")
print(summary)

custom_persona = input("Please describe your interest or the perspective you want to visualize the data from (e.g., 'Someone interested in fuel efficiency'): ")
custom_goals = lida.goals(summary, n=2, persona=custom_persona, textgen_config=textgen_config)
print("\nGenerated Goals based on your persona:")
for i, goal in enumerate(custom_goals):
    print(f"Goal {i+1}: {goal}")

# --- Generate Visualizations based on the first generated goal ---
if custom_goals:
    i = 0
    library = "seaborn" # You can also try "matplotlib", "altair"
    textgen_config_viz = TextGenerationConfig(n=1, temperature=0.2, use_cache=True)
    try:
        charts = lida.visualize(summary=summary, goal=custom_goals[i], textgen_config=textgen_config_viz, library=library)
        if charts:
            print(f"\nGenerated Visualization (using goal '{custom_goals[i]}'):")
            display(charts[0])
        else:
            print("No visualization generated for the first goal.")
    except Exception as e:
        print(f"An error occurred during visualization based on persona goal: {e}")
        print(traceback.format_exc()) # Print detailed traceback for debugging
else:
    print("No goals were generated based on your persona, cannot create visualization.")

# --- Generate visualization via a user query ---
user_query = input("What would you like to see visualized? (eg., 'What is the distribution of car origins?'): ")
textgen_config_query = TextGenerationConfig(n=1, temperature=0.2, use_cache=True)
try:
    charts_query = lida.visualize(summary=summary, goal=user_query, textgen_config=textgen_config_query)
    if charts_query:
        print(f"\nGenerated Visualization (using user query '{user_query}'):")
        display(charts_query[0])
    else:
        print(f"No visualization generated for the query: '{user_query}'.")
except Exception as e:
    print(f"An error occurred during visualization based on user query: {e}")
    print(traceback.format_exc()) # Print detailed traceback for debugging
```
